```python
# This works for 2003
df2010["datestop"] = pd.to_datetime(df2010["datestop"], format="%m%d%Y")
df2010["timestop"] = pd.to_datetime(df2010["timestop"], format="%H:%M", coerce=True)
df2010["hour"] = df2010["timestop"].dt.hour
df2010["minute"] = df2010["timestop"].dt.minute
```

```python
hourly_counts = df2010.groupby(["hour"]).size()#.agg(["mean", "sum"])
hours = hourly_counts.index
counts = hourly_counts
```

```python
plt.bar(hours, counts)
plt.title("Distribution of Stops")
plt.xlabel("Hour of the Day")
plt.ylabel("Count")
plt.xlim([0,24])
plt.xticks([0,4,8,12,16,20,24])
plt.show()
```
