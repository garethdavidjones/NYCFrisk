```python
prec = pd.read_csv("../precinct.csv")
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
check = prec.dropna(0, how='any', subset=["precinct"])
```

```python
agg_prec = check.groupby(["precinct"]).agg("sum").loc[:, "P0010001":"P0020073"]
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(agg_prec)
plt.show()
```
