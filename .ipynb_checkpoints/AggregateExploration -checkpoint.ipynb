```python
import pandas as pd
import matplotlib.pyplot as plt
import glob #Iterating through directory
import imp
import seaborn as sns

%matplotlib inline
```

```python
import SAFutils
imp.reload(SAFutils)
```

```python
# Iterate through files to get an idea of the range of values

# Storage dictionary
holder = {"dayofyear":{}, 
          "weekofyear":{},
          "hour": {}, 
          "monthofyear": {}, 
          "dayofweek": {}
         }

for filename in glob.glob('[0-9]*.csv'):
    print(filename)
    year, df = SAFutils.SAF_data_cleaner(filename)
    
    day_of_week_counts = df.groupby(["dayofweek"]).size()
    hourly_counts = df.groupby(["hour"]).size()
    daily_counts = df.groupby(["dayofyear"]).size()
    week_of_year_counts = df.groupby(["weekofyear"]).size()
    monthly_counts = df.groupby(["month"]).size()
    
    holder["dayofweek"][year] = day_of_week_counts
    holder["hour"][year] = hourly_counts
    holder["dayofyear"][year] = daily_counts
    holder["weekofyear"][year] = week_of_year_counts
    holder["monthofyear"][year] = monthly_counts
    
    print(day_of_week_counts)
    print("{} finished".format(year))
```

```python
import pickle
```

```python
pickle.dump(holder, open("aggregated.p", "wb"))
```

## Daily Stops 

```python
full_day_series = []
for year, data in sorted(holder["dayofyear"].items()):
    full_day_series += list(data)
```

```python
plt.figure(figsize=(20, 5))
plt.plot(full_day_series)
```
